```python
# Реализация функции, которая на вход принимает ключевое слово и даты для поиска,
# а на выходе отдает датафрейм с публикациями в формате: 
# журнал, название статьи, год публикации, количество цитирований. 

import pandas as pd 
import requests
import math
```

```python
# biology, 2023, 2024
keyword = input('Введите ключевое слово: ') 
date_from = int(input('Введите дату начала диапазона поиска: '))
date_to = int(input('Введите дату окончания диапазона поиска: '))
```

```python
def get_pages(keyword, date_from, date_to): 
    global keyword_articles_result

    url = f'https://api.openalex.org/works?filter=abstract.search:{keyword},publication_year:{date_from}-{date_to}'
    response = requests.get(url)
    data = response.json()
    keyword_articles_meta = data['meta']
    keyword_articles_count = int(keyword_articles_meta['count'])
    articles_per_page = int(keyword_articles_meta['per_page'])
    sheets = keyword_articles_count / articles_per_page
    sheets = math.ceil(sheets)
    print(f'Найдено страниц: {sheets}')

    keyword_articles_result = []
    keyword_articles_result = pd.DataFrame(columns=['title', 'publication_year', 'cited_by_count', 'journal_name'])

    # for i in range(1, sheets + 1) вывод всех страниц, их получится много, поэтому выводим первые 5
    for i in range(1, 6):
        url = f'https://api.openalex.org/works?page={i}&filter=abstract.search:{keyword},publication_year:{date_from}-{date_to}'
        response = requests.get(url)
        data = response.json()
        result = []

        for elem in data['results']:
            publication = []
            publication.append(elem['title'])
            publication.append(elem['publication_year'])
            publication.append(elem['cited_by_count'])
            location = elem.get('primary_location')
            if location:
                journal_info = location.get('source')
                if journal_info:
                    journal_name = journal_info.get('display_name')
                    if journal_name:
                        publication.append(journal_name)
                    else:
                        publication.append('-')
                else:
                    publication.append('-')
            else:
                publication.append('-')
            result.append(publication)
        
        pubs = pd.DataFrame(result, columns=['title', 'publication_year', 'cited_by_count', 'journal_name'])
        if pubs.empty:
            continue
        else:
            print(f'\nСТРАНИЦА НОМЕР {i}')
            display(pubs)
            keyword_articles_result = pd.concat([keyword_articles_result, pubs], ignore_index=True)

get_pages(keyword, date_from, date_to)
```
